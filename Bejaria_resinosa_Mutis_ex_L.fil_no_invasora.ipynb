```python
from google.colab import drive
import requests
from PIL import Image
from io import BytesIO
import os
from datetime import datetime
from urllib.parse import urlparse

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)

# Function to create a unique folder name based on the current date and time
def create_unique_folder():
    now = datetime.now()
    timestamp = now.strftime("%Y%m%d%H%M%S")
    return f'Multimedia_{timestamp}'

# Replace <YOUR_GOOGLE_DRIVE_FOLDER_ID> with the actual ID of the Google Drive folder
folder_id = '1Bp1ikrlUICknlQu_0TGna2lxuNYwdG0n'

# Create a unique folder in Google Drive
unique_folder_name = create_unique_folder()
folder_path = os.path.join('/content/drive/MyDrive', folder_id, unique_folder_name)
!mkdir "$folder_path"

# Function to download images and save them to the specified folder
def download_images(urls, folder_path):
    for i, url in enumerate(urls, start=1):
        try:
            response = requests.get(url)
            if response.status_code == 200:
                img = Image.open(BytesIO(response.content))
                img.save(os.path.join(folder_path, f'image_{i}.jpg'))
                print(f"Image {i} downloaded and saved.")
            else:
                print(f"Failed to download image {i}. HTTP status code: {response.status_code}")
        except Exception as e:
            print(f"Error downloading image {i}: {str(e)}")

# Read the multimedia file
multimedia_file_path = '/content/drive/MyDrive/TFM_IMAGENES/Bejaria_resinosa_Mutis_ex_L.fil_no_invasora.txt'
with open(multimedia_file_path, 'r') as file:
    # Skip the header line
    next(file)

    # Extract URLs from the 'identifier' field
    urls = [line.strip().split('\t')[3] for line in file]

# Download and save the images
download_images(urls[:901], folder_path)

print("Images downloaded and saved successfully in the folder:",folder_path)
```
